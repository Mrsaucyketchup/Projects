```python
#max price 35.95
highest_price_pizza = df2.groupby('pizza_name')['unit_price'].max()
highest_price_pizza

```

```python
# prompt: give me only max price and name of of highest_price_pizza

max_price = highest_price_pizza.max()
pizza_with_max_price = highest_price_pizza[highest_price_pizza == max_price].index[0]

print(f"Max Price: {max_price}")
print(f"Pizza with Highest Price: {pizza_with_max_price}")
```

```python
#HW1

least_ordered_pizzas = df2.groupby('pizza_name')['quantity'].sum().sort_values(ascending=True).head(5)
print("5 least ordered pizza types:")
least_ordered_pizzas
```

```python
# prompt: Determine the top 3 most ordered pizza types based on revenue

top_pizza_by_revenue = df2.groupby('pizza_name')['total_price'].sum().sort_values(ascending=False).head(3)

print("Top 3 most ordered pizza types based on revenue:")
top_pizza_by_revenue
```

```python
least_pizza_by_revenue = df2.groupby('pizza_name')['total_price'].sum().sort_values(ascending=True).head(3)

print("least 3 most ordered pizza types based on revenue:")
least_pizza_by_revenue
```

```python
# prompt: draw a histogram for pizza name and quantity
# make it ascending

pizza_quantity = df2.groupby('pizza_name')['quantity'].sum().sort_values(ascending=True)

plt.figure(figsize=(12, 8))
pizza_quantity.plot(kind='bar')
plt.title('Total Quantity Sold per Pizza Name (Ascending)')
plt.xlabel('Pizza Name')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
